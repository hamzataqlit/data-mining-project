Missing value imputation

```python
#Check for duplicate RefId and remove them 
cars.drop_duplicates(subset ="RefId",
                     keep = 'first', inplace = True)
cars
```

```python
#Quantify missing values
cars.isnull().sum()
```

Mean value imputation for all numerical features

```python
cars['WheelTypeID'] = cars['WheelTypeID'].replace({np.nan:cars.WheelTypeID.mean()})
cars['MMRAcquisitionAuctionAveragePrice'] = cars['MMRAcquisitionAuctionAveragePrice'].replace({np.nan:cars.MMRAcquisitionAuctionAveragePrice.mean()})
cars['MMRAcquisitionAuctionCleanPrice'] = cars['MMRAcquisitionAuctionCleanPrice'].replace({np.nan:cars.MMRAcquisitionAuctionCleanPrice.mean()})
cars['MMRAcquisitionRetailAveragePrice'] = cars['MMRAcquisitionRetailAveragePrice'].replace({np.nan:cars.MMRAcquisitionRetailAveragePrice.mean()})
cars['MMRAcquisitonRetailCleanPrice'] = cars['MMRAcquisitonRetailCleanPrice'].replace({np.nan:cars.MMRAcquisitonRetailCleanPrice.mean()})
cars['MMRCurrentAuctionAveragePrice'] = cars['MMRCurrentAuctionAveragePrice'].replace({np.nan:cars.MMRCurrentAuctionAveragePrice.mean()})
cars['MMRCurrentAuctionCleanPrice'] = cars['MMRCurrentAuctionCleanPrice'].replace({np.nan:cars.MMRCurrentAuctionCleanPrice.mean()})
cars['MMRCurrentRetailAveragePrice'] = cars['MMRCurrentRetailAveragePrice'].replace({np.nan:cars.MMRCurrentRetailAveragePrice.mean()})
cars['MMRCurrentRetailCleanPrice'] = cars['MMRCurrentRetailCleanPrice'].replace({np.nan:cars.MMRCurrentRetailCleanPrice.mean()})
```

Frequent occurring value for categorical features

```python
cars.fillna(cars.mode().iloc[6])
cars.fillna(cars.mode().iloc[7])
cars.fillna(cars.mode().iloc[9])
cars.fillna(cars.mode().iloc[11])
```

Feature Engineering

Removing the unnecessary attributes

```python
cars=cars.drop(columns='RefId')
```

```python
cars=cars.drop(columns='WheelTypeID')
```

```python
cars=cars.drop(columns='VehYear')
```

```python
cars=cars.drop(columns='VNZIP1')
```

```python
#Split the purchase date (â€œpurchDateâ€) into three separate features based on date, month and year
PurchDate = cars['PurchDate']
cars[['month','day','year']] =  PurchDate.str.split("/", n=3, expand=True)
cars=cars.drop(columns='PurchDate')
cars
```

```python
#presenting the remaining features and thier types
cars.dtypes
```

```python
#Keeping only cars where WheelType is in (Alloy, Covers)
options = ['Alloy', 'Covers']
cars = cars[cars['WheelType'].isin(options)]
```

```python
#the dataset we have so far
cars
```

```python
#Dummy code the categorical features
cars = pd.get_dummies(cars)
cars
```

```python
#a brief representation of the dataset after the dummy process
cars.dtypes
```

```python
#Performing Standardization to the continuous variables
cols=['VehicleAge','VehOdo','MMRAcquisitionAuctionAveragePrice','MMRAcquisitionAuctionCleanPrice','MMRAcquisitionRetailAveragePrice','MMRAcquisitonRetailCleanPrice','MMRCurrentAuctionAveragePrice','MMRCurrentAuctionCleanPrice','MMRCurrentRetailAveragePrice','MMRCurrentRetailCleanPrice','WarrantyCost','VehBCost']
for col in cols:
    cars[col] = (cars[col] - cars[col].mean()) / cars[col].std()
cars.head()
```

```python
#reporting the new shape of the dataset
print(cars.shape)
```

Applying principal component Analysis(PCA) on this dataset

```python
#importing the necessary decomposition
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(cars)
```

```python
#reporting the captures variance
print(pca.explained_variance_)
```

Applying the classification models

```python
#creating the the target variable and data variable
X = cars.drop(columns='IsBadBuy')
print(X)
y = cars['IsBadBuy']
print(y)
```

```python
#separating X and y into training set and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

Logistic regression model

```python
#importing logistic regression model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state = 0)
model.fit(X_train, y_train)
```

```python
#prediction
y_pred=model.predict(X_test)
y_pred2=model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
#calculating accuracy
score_train = accuracy_score(y_train, y_pred2)
print('\n Score de performance entrainnement:')
print(score_train)

score_test = accuracy_score(y_test, y_pred)
print('\n Score de performance test:')
print(score_test)
```

```python
#calculiting confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print('\nMatrice de confusion :' )
print(cm)
```

SVM model

```python
#for this implementation we used the SVC model
#SVC and SVM are do the same thing but SVC is faster because it supports linear kernel better
from sklearn.svm import LinearSVC
model2 = LinearSVC(random_state=0, tol=1e-5)
model2.fit(X_train, y_train)
```

```python
#prediction
y_pred=model2.predict(X_test)
y_pred2=model2.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
#calculating accuracy
score_train = accuracy_score(y_train, y_pred2)
print('\n Score de performance entrainnement:')
print(score_train)

score_test = accuracy_score(y_test, y_pred)
print('\n Score de performance test:')
print(score_test)
```

```python
#calculiting confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print('\nMatrice de confusion :' )
print(cm)
```
